```python
import time
import datetime
import numpy as np
import pandas as pd
```

## increasing the time of pulses and LP 

```python
WELL='A1'

'start the variables'
laser_power = []
time_turn_ON = []
time_turn_OFF = []
time_laser_ON = []
ON_time = []
repeat = []
time_ON_extraction = []

laser.off()

starting_time = 0.01 #laser ON starting in seconds (10ms)


for t in range (15):  #in total we make 4 measurements at different times of exposure to the laser
    
    print(starting_time)
    
    current_power = 0.02 #the starting current power of the laser in W
    
    while current_power <= 0.1:  #we go til 100mW = 0.1W
        
            repetition = 1 #we start from repetiotion1, we want to make 10
        
            for r in range (3): #number of repetitions we are making of each condition
                
                laser.set_power(current_power)
                print(current_power) 

                laser.on()

                nowON=datetime.datetime.now()
                now_ON = datetime.datetime.time(nowON)
                print(now_ON)   # print when the laser turns ON
                
                time.sleep(starting_time)
                
                laser.off()

                nowOFF = datetime.datetime.now()
                now_OFF = datetime.datetime.time(nowOFF)
                print(now_OFF)

                time.sleep(2)  #OFF always same time, 20sec
                
                #time_ON_extraction = np.array(time_turn_OFF)-np.array(time_turn_OFF)
                time_ON_extraction = nowOFF- nowON

                ON_time.append(int(starting_time*1000))
                laser_power.append(int(current_power*1000))
                time_turn_ON.append(now_ON)
                time_turn_OFF.append(now_OFF) 
                repeat.append(int(repetition))  #las repeticiones
                time_laser_ON.append(time_ON_extraction)
                
                repetition += 1
        
            current_power += 0.02  #we increase in 20mW after each 10 repetitions
        
    starting_time += 0.01 #increase in 10msec the laser ON time for the experiment
    
experiment = pd.DataFrame({'Repetition':repeat, 'ON duration (ms)':ON_time,'Laser Power (mW)':laser_power,
                           'Turn ON Laser Time':time_turn_ON, 'Turn OFF Laser Time':time_turn_OFF,
                           'Turn ON Laser Time (ms)':time_laser_ON} ) #a dataframe is created to save the times of ON/OFF of the experiment

now = datetime.datetime.now()
date = '%s0%s%s' % (now.year, now.month, now.day)
experiment.to_csv(f'{date}_MEAs_{WELL}.csv')
experiment
```
